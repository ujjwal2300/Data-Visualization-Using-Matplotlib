```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Most of the Matplotlib utilities lies under the pyplot subModule
```

```python
#Line Plot

x_points = np.array([2,4])
y_points = np.array([5,10])

fig = plt.figure(figsize = (3,3))
plt.plot(x_points, y_points)
plt.show()
```

```python
x_points = np.array([2,4])
y_points = np.array([5,10])

fig = plt.figure(figsize = (3,3))
plt.plot(x_points, y_points,'o--b') #marker = o #line = -- #color = blue
plt.show()
```

```python
x_points = np.array([2,4])
y_points = np.array([5,10])

fig = plt.figure(figsize = (3,3))
plt.plot(x_points, y_points,'o--b')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Using Matplotlib")
plt.show()
```

```python
#Scatter Plot

x1 = [10, 14, 45, 32, 23]
y1 = [21, 45, 3, 35, 29]

fig = plt.figure(figsize = (3,3))
plt.scatter(x1, y1, c ='red',
           linewidth = 2,
           marker = 'o', #this default marker
           edgecolor = 'green',
           s = 50)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot using matplotlib")
plt.show()
```

```python
#Two Scatter Plots overlapped

#dataset-1
x1 = [10, 14, 45, 32, 23]
y1 = [21, 45, 3, 35, 29]

#dataset-2
x2 = [66, 72, 80, 98, 74]
y2 = [90, 56, 88, 74, 69]

fig = plt.figure(figsize = (3,3))
plt.scatter(x1, y1, c ='red',
           linewidth = 2,
           marker = 'o', #this default marker
           edgecolor = 'green',
           s = 50)


plt.scatter(x2, y2, c ='yellow',
           linewidth = 2,
           marker = '^', #this default marker
           edgecolor = 'red',
           s = 200)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot using matplotlib")
plt.show()
```

```python
#Bar Plot using Matplotlib
#Average Salary Index

data = {'Data Scientist':10, 'Software Engineer':6.5, 'Software Testing':5,
       'Data Analyst':7.5,'Data Engineer':7}

job_profile = list(data.keys())
salary_index = list(data.values())

fig = plt.figure(figsize = (4,2))

#creating our bar plot
plt.bar(job_profile, salary_index, color='blue',width = 0.3)
plt.xticks(rotation=90, ha='right') #Rotate labels 45 degree to the right
plt.xlabel("Job Profile")
plt.ylabel("Salary index")
plt.title("Job Profile vs Salary index")
plt.show()
```

```python
data = np.random.normal(loc = 200, scale = 20, size = 250)
#loc: mean, scale: standard deviation, size = number of samples

fig = plt.figure(figsize = (3,3))
plt.hist(data, color ='blue', bins = 10)
plt.xlabel("data value")
plt.ylabel("frequency")
plt.title("Histogram using Matplotlib")

plt.show()
```

```python
# Pie Chart using Matplotlib

#Revenue share in E-Commerce
data = {'Clothes':50, 'Househole Device': 20, 'Baby Products':15,
       'Grocery':13, 'Mobile':9, 'Laptop':6}

items = list(data.keys())
percentage_share = list(data.values())

fig = plt.figure(figsize = (5,5))
plt.pie(percentage_share, labels = items, startangle = 90, counterclock=False)
plt.legend(loc = 'upper right',bbox_to_anchor=(1, 0, 0.5,1))
plt.show()
```
